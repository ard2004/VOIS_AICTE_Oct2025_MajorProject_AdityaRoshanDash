## Analyze genre popularity change over time

### Subtask:
Group the DataFrame by year and genre to count the number of releases for each genre in each year, select the top N most common genres and visualize their release trends over the years using line plots to see how their popularity has changed.


**Reasoning**:
Create a new DataFrame by grouping the original DataFrame `df` by 'years' and 'Type', and counting the number of releases for each group. Then reset the index and name the count column appropriately. After that, split the 'Type' column in the new DataFrame into individual genres. Finally create a new DataFrame with columns for 'year', 'genre', and 'count' where each row represents the number of releases for a specific genre in a given year. Identify the top 10 most frequent genres from the `genre_counts` Series and filter the DataFrame created to include only the top 10 genres.



```python
release_by_year_type = df.groupby(['years', 'Type']).size().reset_index(name='count')
genre_release_trend = []
for index, row in release_by_year_type.iterrows():
    genres = row['Type'].split(', ')
    for genre in genres:
        genre_release_trend.append({'years': row['years'], 'genre': genre, 'count': row['count']})
genre_release_trend_df = pd.DataFrame(genre_release_trend)
top_10_genres = genre_counts.head(10).index.tolist()
top_genre_release_trend_df = genre_release_trend_df[genre_release_trend_df['genre'].isin(top_10_genres)]
```

**Reasoning**:
Create a line plot to visualize the trend of releases for each of the top 10 genres over the years, add a title and axis labels to the plot, and display the plot.



```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='years', y='count', hue='genre', data=top_genre_release_trend_df)
plt.title('Release Trend of Top 10 Genres Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```
